```python
with open("data/raw/riowomen800mfreestyle.txt",'r') as input:
    raw = input.read()
```

```python
#We extract infor per swimmer
lines = raw.split("\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
```

```python
def extractSwimmerData(lines):
    pieces = lines[0].split(" ")
    swimmerName = pieces[3]+" "+pieces[2]
    swimmerTimes = []
    swimmerTimes.append(lines[1].split(" ")[2])
    for i in range(15):
        swimmerTimes.append(lines[2+i*2])
    swimmerTimes = map(lambda e: float(e),swimmerTimes)
    result=[]
    for i in range(len(swimmerTimes)):
        result.append({'name': swimmerName,'time':swimmerTimes[i],'length':50*(i+1)})
    return result
```

```python
#Extract all the Swimmers Data
result = []
for i in range(len(lines)/32):
    result+=extractSwimmerData(lines[i*31:(i+1)*31])
result=pd.DataFrame(result)
```

```python
result[:5]
```

```python
##http://blog.insightdatalabs.com/advanced-functionality-in-seaborn/
sns.factorplot(data=result, x="length", y="time", col="name",col_wrap=3)
```

```python
#Let's see the results for the US National team Classification
with open("data/raw/women800mUSNat2008.txt") as input:
    lines = input.readlines()
```

```python
def extract(lines):
    pieces = lines[0].split(" ")
    name = pieces[3] + " " + pieces[2][:-1]
    times = []
    for line in lines[1:]:
        e = line.replace("  ","").replace("(","").replace(")","").replace("\n","").split(" ")
        #We are in the first line
        if len(e)%2!=0:
            times+=[float(e[i]) for i in range(len(e)) if i%2==0]
        else:
            times+=[float(e[i]) for i in range(len(e)) if i%2!=0]
    result=[]
    for i in range(len(times)):
        result.append({'name': name,'time':times[i],'length':50*(i+1)})
    return result
```

```python
#Extract all the Swimmers Data
result = []
for i in range((len(lines)-1)/5):
    result+=extract(lines[i*5+1:(i+1)*5+1])
result=pd.DataFrame(result)
```

```python
sns.factorplot(data=result, x="length", y="time", col="name",col_wrap=3)
```
